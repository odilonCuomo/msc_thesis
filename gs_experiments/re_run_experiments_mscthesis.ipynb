##Asymmetry in props

```python
args = utils.args.Args()
args.n = 15
ticks = 40
tick_range = range(ticks + 1)
dispersion_range = [i / ticks for i in tick_range]
args.num_runs = 5000

#iterate over
noisy_sides = ["suitors", "reviewers"]
other_side_phi = [i / 5 for i in range(6)]
borda_poss = [True, False]
path = '/content/drive/MyDrive/Uni/Oxford/ACS/thesis/diversity_matching_pb/results/rerun1/n15/asymmetry/'

for side in noisy_sides:
  for phi in other_side_phi:
    for borda in borda_poss:
      args.noisy_side = side
      args.other_phi = phi
      stats, borda_stats_sui, borda_stats_rev = asymmetry.asymmetry_prop_borda(args.n, args.num_runs, dispersion_range, args.other_phi, noisy_side=args.noisy_side, borda=borda)
      path = os.path.join(path, "borda" if borda else "props")
      path = os.path.join(path, noisy_sides[0])

      sub = ''
      for s in path.split('/'):
        sub += s + '/'
        if not os.path.isdir(sub):
          os.mkdir(sub)
      file_name = utils.args.get_path_name(args)
      file_path = os.path.join(path, file_name)

      if not borda:
          graph_mean_stdev(stats, file_name, file_path, "phi", "E[T_n]")
      else:
          graph_mean_stdev_multiple([borda_stats_sui, borda_stats_rev], file_name, file_path, "phi", "Mean Borda score", ["suitors borda", 'reviewers borda'])
```
